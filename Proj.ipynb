# Projeto de Otimização 

Bibliotecas utilizadas:

```python
import pulp as plp
import pandas as pd
```

Seleção do número de variáveis:

```python
num_vars = int(input("Numero de variavies do problema (2,3,4): "))
while num_vars not in (2, 3, 4):
    print("Número inválido (2,3,4)")
    num_vars = int(input("Numero de variavies do problema (2,3,4): "))
```

Seleção do tipo de problema (max ou min):

```python
f_obj = input("Função objetivo (max ou min): ")
while f_obj not in ("max", "min"):
    print("Opcao invalida")
    f_obj = input("Função objetivo (max ou min): ")

if f_obj == "max":
    z = plp.LpProblem ("Maximizar", plp.LpMaximize)

elif f_obj == "min":
    z = plp.LpProblem ("Minimizar", plp.LpMinimize)
```

Resolvendo o problema: 

```python
if num_vars == 2:
    a = plp.LpVariable('a', lowBound=0, cat='real')
    b = plp.LpVariable('b', lowBound=0, cat='real')

    c_a = float(input("Coeficiente da variavel A: "))
    c_b = float(input("Coeficiente da variavel B: "))
    
    #F_objetiva
    z += c_a*a + c_b*b

    flag_rest = int(input("Adicionar restricao (1 - sim, 2 - nao): "))
    while flag_rest not in (1,2):
        print("Opcao invalida")
        flag_rest = int(input("Adicionar restricao (1 - sim, 2 - nao): "))
    while flag_rest == 1:
        c_a = float(input("Coeficiente da variavel A: "))
        c_b = float(input("Coeficiente da variavel B: "))

        op = input("Sinal (>=, <=, ==): ")
        while op not in (">=", "<=", "=="):
            print("Opcao invalida")
            op = input("Sinal (>=, <=, ==): ")

        comp = float(input("Valor de comparacao: "))

        if op == ">=":
            z +=  a*c_a + b*c_b >= comp
            
        elif op == "<=":
            z +=  a*c_a + b*c_b <= comp   

        elif op == "==": 
            z +=  a*c_a + b*c_b == comp

        flag_rest = int(input("Adicionar restricao (1 - sim, 2 - nao): "))
        while flag_rest not in (1,2):
            print("Opcao invalida")
            flag_rest = int(input("Adicionar restricao (1 - sim, 2 - nao): "))
    
    z.solve()
    print("Valores otimos de A: ", a.varValue)
    print("Valores otimos de B: ", b.varValue)
    print("Valor da função objetivo:", plp.value(z.objective))

    o = [{'Nome':name,'Preco-sombra':c.pi,'Sobra': c.slack} for name, c in z.constraints.items()]
    print(pd.DataFrame(o))


elif num_vars == 3:
    a = plp.LpVariable('a', lowBound=0, cat='real')
    b = plp.LpVariable('b', lowBound=0, cat='real')
    c = plp.LpVariable('c', lowBound=0, cat='real')

    c_a = float(input("Coeficiente da variavel A: "))
    c_b = float(input("Coeficiente da variavel B: "))
    c_c = float(input("Coeficiente da variavel C: "))
    
    #F_objetiva
    z += c_a*a + c_b*b + c_c*c

    flag_rest = int(input("Adicionar restricao (1 - sim, 2 - nao): "))
    while flag_rest not in (1,2):
        print("Opcao invalida")
        flag_rest = int(input("Adicionar restricao (1 - sim, 2 - nao): "))
    while flag_rest == 1:
        c_a = float(input("Coeficiente da variavel A: "))
        c_b = float(input("Coeficiente da variavel B: "))
        c_c = float(input("Coeficiente da variavel C: "))

        op = input("Sinal (>=, <=, ==): ")
        while op not in (">=", "<=", "=="):
            print("Opcao invalida")
            op = input("Sinal (>=, <=, ==): ")

        comp = float(input("Valor de comparacao: "))

        if op == ">=":
            z +=  a*c_a + b*c_b + c*c_c >= comp
            
        elif op == "<=":
            z +=  a*c_a + b*c_b + c*c_c <= comp   

        elif op == "==": 
            z +=  a*c_a + b*c_b + c*c_c == comp

        flag_rest = int(input("Adicionar restricao (1 - sim, 2 - nao): "))
        while flag_rest not in (1,2):
            print("Opcao invalida")
            flag_rest = int(input("Adicionar restricao (1 - sim, 2 - nao): "))

    z.solve()
    print("Valores otimos de A: ", a.varValue)
    print("Valores otimos de B: ", b.varValue)
    print("Valores otimos de C: ", c.varValue)
    print("Valor da função objetivo:", plp.value(z.objective))

    o = [{'Nome':name,'Preco-sombra':c.pi,'Sobra': c.slack} for name, c in z.constraints.items()]
    print(pd.DataFrame(o))


elif num_vars == 4:
    a = plp.LpVariable('a', lowBound=0, cat='real')
    b = plp.LpVariable('b', lowBound=0, cat='real')
    c = plp.LpVariable('c', lowBound=0, cat='real')
    d = plp.LpVariable('d', lowBound=0, cat='real')

    c_a = float(input("Coeficiente da variavel A: "))
    c_b = float(input("Coeficiente da variavel B: "))
    c_c = float(input("Coeficiente da variavel C: "))
    c_d = float(input("Coeficiente da variavel D: "))
    
    #F_objetiva
    z += c_a*a + c_b*b + c_c*c + c_d*d

    flag_rest = int(input("Adicionar restricao (1 - sim, 2 - nao): "))
    while flag_rest not in (1,2):
        print("Opcao invalida")
        flag_rest = int(input("Adicionar restricao (1 - sim, 2 - nao): "))
    while flag_rest == 1:
        c_a = float(input("Coeficiente da variavel A: "))
        c_b = float(input("Coeficiente da variavel B: "))
        c_c = float(input("Coeficiente da variavel C: "))
        c_d = float(input("Coeficiente da variavel D: "))

        op = input("Sinal (>=, <=, ==): ")
        while op not in (">=", "<=", "=="):
            print("Opcao invalida")
            op = input("Sinal (>=, <=, ==): ")

        comp = float(input("Valor de comparacao: "))

        if op == ">=":
            z +=  a*c_a + b*c_b + c*c_c + d*c_d >= comp
            
        elif op == "<=":
            z +=  a*c_a + b*c_b + c*c_c + d*c_d <= comp   

        elif op == "==": 
            z +=  a*c_a + b*c_b + c*c_c + d*c_d == comp

        flag_rest = int(input("Adicionar restricao (1 - sim, 2 - nao): "))
        while flag_rest not in (1,2):
            print("Opcao invalida")
            flag_rest = int(input("Adicionar restricao (1 - sim, 2 - nao): "))
    
    z.solve()
    print("Valores otimos de A: ", a.varValue)
    print("Valores otimos de B: ", b.varValue)
    print("Valores otimos de C: ", c.varValue)
    print("Valores otimos de D: ", d.varValue)
    print("Valor da função objetivo:", plp.value(z.objective))

    o = [{'Nome':name,'Preco-sombra':c.pi,'Sobra': c.slack} for name, c in z.constraints.items()]
    print(pd.DataFrame(o))

    
```
